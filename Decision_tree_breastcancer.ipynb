#### Testing, Training and Label Encoding -

```python
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()
data.species = lc.fit_transform(data.Class)
```

```python
x = data[['Code', 'Clump', 'UniCell_Size', 'Uni_CellShape', 'MargAdh','SEpith','BareN','BChromatin', 'NoemN','Mitoses']]
y = data['Class']
```

```python
data.Class[: 3]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

#### Decision Tree - 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

```python
y_pred = dtree.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
print('Classification Matrix:\n',  classification_report(y_test, y_pred))
print('\nConfusion Matrix:\n',  confusion_matrix(y_test, y_pred))
print('\nAccuracy:',  accuracy_score(y_test, y_pred))
```

```python
#Decision Tree -

from sklearn.tree import plot_tree
plt.figure(figsize =(20,20))
d = plot_tree(decision_tree = dtree, feature_names = data.columns,
                   class_names = ['Benign', ' Malignant'], filled = True,
                        precision = 4, rounded = True)
```
